```python
#huấn luyện sử dụng Logistic Regression
from sklearn.linear_model import LogisticRegression
  
model = LogisticRegression()
model.fit(x_train, y_train)
  
accuracy_score(y_train, model.predict(x_train)), accuracy_score(y_test, model.predict(x_test))
```

```python
#huấn luyện sử dụng Decision Tree
from sklearn.tree import DecisionTreeClassifier
  
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
  
accuracy_score(y_train, model.predict(x_train)), accuracy_score(y_test, model.predict(x_test))
```

```python
#ma trận nhầm lẫn của cây quyết định
from sklearn import metrics
import matplotlib.pyplot as plt
cm = metrics.confusion_matrix(y_test, model.predict(x_test))
  
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])
  
cm_display.plot()
plt.show()
```

```python
#do cây quyết định có tỉ lệ dự đoán đúng cao hơn nên ta sẽ sử dụng cây quyết định để dự đoán
for i in range(x_test.shape[0]):
    prediction = model.predict(x_test[i])
    if(prediction[0] == 0):
        print('Tin thứ {} là tin tức giả'.format(i + 1))
    else:
        print('Tin thứ {} là tin tức thật'.format(i + 1))
```
